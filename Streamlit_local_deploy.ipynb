```python
%%writefile app.py
import numpy as np
import pandas as pd
import streamlit as st
from sklearn.linear_model import LogisticRegression
import pickle
from pickle import load

st.title('Model Deployment: Logistic Regression')

st.sidebar.header('User Input Parameters')

# Function to preprocess the input data
def preprocess_input_data(sex, age, pclass, sibsp, parch, fare, embarked):
    sex_map = {'male': 0, 'female': 1}
    embarked_map = {'C': 1, 'Q': 2, 'S': 0}

    sex = sex_map.get(sex.lower(), 0)
    embarked = embarked_map.get(embarked.upper(), 0)

    # One-hot encode the embarked column (C, Q, S) into Embarked_C, Embarked_Q, Embarked_S
    embarked_Q = 1 if embarked == 2 else 0
    embarked_S = 1 if embarked == 0 else 0

    # Create the input array with all 8 features
    input_data = np.array([sex, age, pclass, sibsp, parch, fare, embarked_Q, embarked_S]).reshape(1, -1)
    return input_data

# User input fields
sex = st.selectbox('Sex', ['Male', 'Female'])
age = st.number_input('Age', min_value=0, max_value=100, value=30)
pclass = st.selectbox('Pclass', [1, 2, 3])
sibsp = st.number_input('SibSp (Number of Siblings/Spouses Aboard)', min_value=0, value=0)
parch = st.number_input('Parch (Number of Parents/Children Aboard)', min_value=0, value=0)
fare = st.number_input('Fare', min_value=0, value=50)
embarked = st.selectbox('Embarked', ['C', 'Q', 'S'])

# Load the trained model
with open('titanic_model.pkl', 'rb') as f:
    model = pickle.load(f)

# When the user clicks the 'Predict' button
if st.button('Predict'):
    input_data = preprocess_input_data(sex, age, pclass, sibsp, parch, fare, embarked)

    # Get the predicted probabilities for each class
    prediction_proba = model.predict_proba(input_data)

    # The model returns probabilities for both classes: [no_survival, survival]
    survival_probability = prediction_proba[0][1]

    # Display the survival prediction based on the probability threshold (e.g., 0.5)
    if survival_probability > 0.5:
        st.write(f"Survival prediction: Yes (Probability: {survival_probability:.2f})")
    else:
        st.write(f"Survival prediction: No (Probability: {survival_probability:.2f})")


```
