Exercise 1: Prime Numbers

```python
# Function to check if a number is prime
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True

# Input from user
number = int(input("Enter a number to check if it is prime: "))
if is_prime(number):
    print(f"{number} is a prime number.")
else:
    print(f"{number} is not a prime number.")
```

Exercise 2: Product of Random Numbers


```python
import random

# Generate two random numbers
num1 = random.randint(1, 10)
num2 = random.randint(1, 10)

# Ask the user to calculate the product
print(f"Calculate the product of {num1} and {num2}.")
user_answer = int(input("Enter your answer: "))

# Calculate the correct product
correct_product = num1 * num2

# Check if the answer is correct
if user_answer == correct_product:
    print("Correct! Well done.")
else:
    print(f"Incorrect. The correct product is {correct_product}.")
```

Exercise 3: Squares of Even/Odd Numbers


```python
# Input range and choice from the user
start = int(input("Enter the start of the range: "))
end = int(input("Enter the end of the range: "))
choice = input("Do you want to calculate squares of 'even' or 'odd' numbers? ").strip().lower()

# Validate choice and calculate squares
if choice == 'even':
    print(f"Squares of even numbers between {start} and {end}:")
    for num in range(start, end + 1):
        if num % 2 == 0:
            print(f"{num}^2 = {num**2}")
elif choice == 'odd':
    print(f"Squares of odd numbers between {start} and {end}:")
    for num in range(start, end + 1):
        if num % 2 != 0:
            print(f"{num}^2 = {num**2}")
else:
    print("Invalid choice. Please enter 'even' or 'odd'.")
```

Exercise 4: Word Counter


```python
# Input text from the user
input_text = str(input("Enter a text: "))

# Split the text into words
words = input_text.split()

# Count word frequencies
word_count = {}
for word in words:
    word = word.lower().strip(".,")  # Convert to lowercase and remove punctuation
    word_count[word] = word_count.get(word, 0) + 1

total_words= sum(word_count.values())

print(f"Total number of words: {total_words}")

# Print the word counts
print("Word frequencies:")
for word, count in word_count.items():
    print(f"'{word}': {count}")
```

Exercise 5: Check for Palindrome


```python
# Function to check if a string is a palindrome
def is_palindrome(s):
    # Remove spaces, punctuation, and convert to lowercase
    s = ''.join(filter(str.isalnum, s)).lower()
    return s == s[::-1]

# Input from user
input_string = input("Enter a string to check if it is a palindrome: ")

# Check and print result
if is_palindrome(input_string):
    print(f"'{input_string}' is a palindrome.")
    print(f"True")
else:
    print(f"'{input_string}' is not a palindrome.")
```
